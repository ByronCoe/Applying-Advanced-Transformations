```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
# Load the data from CSV files
superhero_info = pd.read_csv('superhero_info - superhero_info.csv')
superhero_powers = pd.read_csv('superhero_powers - superhero_powers.csv')
```

```python
superhero_info.info()
```

```python
superhero_info.head()
```

```python
superhero_powers.info()
```

```python
superhero_powers.head()
```

```python
# Merge the two dataframes on 'Hero' column
df = pd.concat((superhero_powers, superhero_info), axis = 1)
df
```

```python
df['Hero|Publisher'].str.split('|',expand=True)
```

```python
df[['Hero','Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
df.head(2)
```

```python
## drop the original column 
df = df.drop(columns=['Hero|Publisher', 'Hero'])
df.head(2)
```

```python
df.rename(columns={'hero_names': 'Hero'}, inplace=True)
df.head(2)
```

```python
df['Measurements'].str.split(',',expand=True)
```

```python
df[['Height(cm)','Weight(kg)']] = df['Measurements'].str.split(',',expand=True)
df.head(2)
```

```python
## drop the original column 
df = df.drop(columns=['Measurements'])
df.head(2)
```

```python
to_replace = ['{','}',"'Height': ","'",' cm']

for char in to_replace:
    df['Height(cm)'] = df['Height(cm)'].str.replace(char,'',regex=False)
    
df['Height(cm)'].head()
```

```python
to_replace = ['{','}',"'Weight': ","'",' kg']

for char in to_replace:
    df['Weight(kg)'] = df['Weight(kg)'].str.replace(char,'',regex=False)
    
df['Weight(kg)'].head()
```

```python
df['Weight(kg)'] = pd.to_numeric(df['Weight(kg)'], errors='coerce')
df['Weight(kg)'].head()
```

```python
df['Height(cm)'] = pd.to_numeric(df['Height(cm)'], errors='coerce')
df['Height(cm)'].head()
```

```python
df['Powers'].value_counts()
```

```python
# Split the 'Powers' column into separate columns for each power and one-hot encode them
powers_dummies = df['Powers'].str.get_dummies(sep=',')
powers_dummies
```

```python
# Concatenate the one-hot encoded powers with the original DataFrame
df = pd.concat([df, powers_dummies], axis=1)
```

```python
df.drop(columns=['Powers'], inplace=True)
df
```

```python
# Calculate the average weight of superheroes with Super Speed
average_weight_with_super_speed = df[df['Super Speed'] == 1]['Weight(kg)'].mean()

# Calculate the average weight of superheroes without Super Speed
average_weight_without_super_speed = df[df['Super Speed'] == 0]['Weight(kg)'].mean()

print("Average weight of superheroes with Super Speed in kg:", average_weight_with_super_speed)
print("Average weight of superheroes without Super Speed in kg:", average_weight_without_super_speed)
```

```python
# Calculate average height of heroes for each publisher
average_height_by_publisher = df.groupby('Publisher')['Height(cm)'].mean()

print("Average height of heroes by publisher in cm:")
print(average_height_by_publisher)
```
